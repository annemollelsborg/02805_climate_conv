```python
import os
from dotenv import load_dotenv
import requests

load_dotenv()
api_key = os.getenv("NEWS_API_KEY")

url = "https://newsapi.org/v2/everything"
params = {
    "q": '"climate change" OR #climatechange OR #climate OR "global warming" OR climate OR carbon OR emissions', ## same keywords than X tweets
    "from": "2025-11-06", ## WARNING: Basic Free plan of News Api only permits us to search for article a month from current date
    "to": "2018-11-17",
    "language": "en",
    "sortBy": "publishedAt",
    "apiKey": api_key
}
```

```python
## get the articles
response = requests.get(url, params=params)
articles = response.json()["articles"]
```

```python
## Basic stats
for a in articles:
    print(a["title"], "-", a["publishedAt"])
names = []
authors = []
for a in articles:
    names.append(a["source"]["name"])
    authors.append(a["author"]) ## regex here
names_set = set(names)
authors_set = set(authors)
print(names_set)
print(len(names_set))

print(authors_set)
print(len(authors_set))
len(articles)
```
